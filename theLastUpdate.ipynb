```python
#              BB1 x1   y1     x2   y2    BB2 x1   y1     x2   y2    BB3 x1   y1     x2   y2     BB4 x1   y1     x2   y2
boundryBoxes = [[[400, 320], [650,550]], [[1200, 550] , [1400,770]],  [[690, 880] , [800,1000]],  [[1050, 130] , [1200,240]]]
```

```python
len(boundryBoxes)
```

```python
def colors(class_id , id):
  color = (255,0,0) # def green
  #form coco classes
  # 2: car
  # 3: motorcycle
  # 7: truck
  if class_id == '2':
    color = (0,255,0)
  elif class_id == '3':
    color = (0,0,255)
  elif class_id == '7':
    color = (255,255,0)

  if id in ides:
    color = (0,0,0)
  return color
```

```python
dectCount = {
    '0': 0,
    '1': 0,
    '2': 0,
    '3': 0
}
```

```python
dectCount
```

```python
ides = set()
```

```python
def check(center , id):

  if 400 < center[0] < 650 and 320 < center[0] < 550:
    if id not in ides:
        ides.add(id)
        dectCount.update({str(0): dectCount.pop(str(0)) + 1})
        print("count++")
  elif 1200 < center[0] < 1400 and 550 < center[0] < 770:
      if id not in ides:
        ides.add(id)
        dectCount.update({str(1): dectCount.pop(str(1)) + 1})
        print("count++")
  elif 690 < center[0] < 800 and 880 < center[0] < 1000:
    if id not in ides:
        ides.add(id)
        dectCount.update({str('2'): dectCount.pop('2') + 1})
        print("count++")
  elif 1050 < center[0] < 1200 and 130 < center[0] < 240:
    if id not in ides:
        ides.add(id)
        dectCount.update({str('3'): dectCount.pop('3') + 1})
        print("count++")
```

```python
thickness = 2
font = cv.FONT_HERSHEY_SIMPLEX
font_scale = 0.5
font_thickness = 1

ret = True
while ret:
    ret, frame = cap.read()
    if not ret:
        break
    results = model.track(frame, persist=True)
    for result in results:
        for obj in result.boxes:
            bbox = obj.xyxy[0].cpu().numpy()
            class_id = int(obj.cls[0].cpu().numpy()) if obj.cls is not None else -1
            conf = obj.conf[0].cpu().numpy() if obj.conf is not None else 0.0
            obj_id = int(obj.id[0].cpu().numpy()) if obj.id is not None else -1
            x1, y1, x2, y2 = map(int, bbox)
            xc = (x1 + x2) // 2
            yc = (y1 + y2) // 2
            center = [xc , yc]
            check(center , obj_id)
            cv.rectangle(frame, (x1, y1), (x2, y2), colors(class_id,obj_id), thickness)
            cv.rectangle(frame, (xc + 1 , yc - 1), (xc + 1, yc + 1), colors(class_id,obj_id), thickness)
            label = f'ID: {obj_id} Class: {class_id} Conf: {conf:.2f}'
            cv.putText(frame, label, (x1, y1 - 10), font, font_scale, colors(class_id,obj_id), font_thickness, cv.LINE_AA)
            for i in range(4):
              cv.rectangle(frame, (boundryBoxes[i][0][0], boundryBoxes[i][0][1]), (boundryBoxes[i][1][0], boundryBoxes[i][1][1]), colors(class_id,obj_id), thickness)
              label1 = f'Car caunt: {dectCount.get(str(i))}'
              cv.putText(frame, label1, (boundryBoxes[i][0][0], boundryBoxes[i][0][1] - 10), font, font_scale, colors(class_id,obj_id), 3 , cv.LINE_AA)


    output.write(frame)

cap.release()
output.release()
```
